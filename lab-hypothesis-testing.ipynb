- We posit that Legendary Pokemons have different stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary. Choose the propper test and, with 5% significance, comment your findings.


```python
columns = ["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]
legendary = df[df["Legendary"]==True]
nonlegendary = df[df["Legendary"]==False]
```

```python
results = {}
for item in columns:
    t_stat2, p_value2 = st.ttest_ind(legendary[item],nonlegendary[item], equal_var=False)
    results[item] = {"t-statistic": t_stat2, 
                     "p-value": p_value2}
```

```python
alpha = 0.05  # Significance level
for item, result in results.items():
    print(f"\nStat: {item}")
    print(f"t-statistic: {result['t-statistic']}")
    print(f"p-value: {result['p-value']}")
    
    # Decision
    if result['p-value'] < alpha:
        print(f"Reject the null hypothesis: There is a significant difference in {item} between Legendary and Non-Legendary PokÃ©mon.")
    else:
        print(f"Fail to reject the null hypothesis: No significant difference in {item} between Legendary and Non-Legendary PokÃ©mon.")
```

**We posit that houses close to either a school or a hospital are more expensive.**

- School coordinates (-118, 37)
- Hospital coordinates (-122, 34)

We consider a house (neighborhood) to be close to a school or hospital if the distance is lower than 0.50.

Hint:
- Write a function to calculate euclidean distance from each house (neighborhood) to the school and to the hospital.
- Divide your dataset into houses close and far from either a hospital or school.
- Choose the propper test and, with 5% significance, comment your findings.
 

```python
school_coordinates = (-118, 37)
hospital_coordinates = (-122, 34)
```

```python
def calc_dist(lat1, lon1, lat2, lon2):
    return np.sqrt((lat1 - lat2)**2 + (lon1 - lon2)**2)
```

```python
df['distance_to_school'] = calc_dist(df['longitude'], df['latitude'], school_coordinates[0], school_coordinates[1])
df['distance_to_hospital'] = calc_dist(df['longitude'], df['latitude'], hospital_coordinates[0], hospital_coordinates[1])
```

```python
df['close_to_school_or_hospital'] = (df['distance_to_school'] < 0.50) | (df['distance_to_hospital'] < 0.50)
```

```python
df
```

```python
close = df[df['close_to_school_or_hospital'] == True]['median_house_value']
far = df[df['close_to_school_or_hospital'] == False]['median_house_value']
```

```python
t_stat, p_value = st.ttest_ind(close, far, equal_var=False)
print(f"EstadÃ­stico t: {t_stat}")
print(f"Valor p: {p_value}")
```
